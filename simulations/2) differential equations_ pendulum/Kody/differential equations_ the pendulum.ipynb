Solving differential equations: the pendulum - Euler method (3.74)


```python
import math
import matplotlib.pyplot as plt
import numpy as np

theta_0 = 2*math.pi / 3
V_theta_0 = 0
g = 9.8
L = 1

t_step = 0.001 #delta
N = 100000

def acceleration(theta):
    acc = -g/L*math.sin(theta)
    return acc

acc_0 = acceleration(theta_0)

position_results = [theta_0]
velocity_results = [V_theta_0] 
acceleration_results = [acc_0]


for t in range(N): #because in the beginning we have initial conditions *
    
    theta = position_results[-1] + velocity_results[-1] * t_step #new position
    velocity = velocity_results[-1] + acceleration_results[-1] * t_step #new velocity
    acc = acceleration(theta) #with new position (theta)
    
    position_results.append(theta)
    velocity_results.append(velocity)
    acceleration_results.append(acc)


#T = list(map(lambda x: x * t_step, range(N+1))) #* thats why here we have N+1

# list comprehension: standard Python
#T = [x * t_step for x in range(N+1)]

# from numpy: arange or linspace
T = np.arange(N+1) * t_step

plt.plot(T, velocity_results)
plt.xlabel("T [s]")
plt.ylabel("V [rad]")
plt.title("$\\dot\\theta$ (s)")
plt.show()


plt.plot(T, position_results)
plt.xlabel("T [s]")
plt.ylabel("$\\theta$ [rad]")
plt.title("$\\theta$ (s)")
plt.grid()
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

theta_0 = 2*math.pi / 3
V_theta_0 = 0
g = 9.8
L = 1

t_step = 0.01 #delta
N = 10000

def acceleration(theta):
    acc = -g/L*math.sin(theta)
    return acc

acc_0 = acceleration(theta_0)

position_results = [theta_0]
velocity_results = [V_theta_0] 
acceleration_results = [acc_0]


for t in range(N): #bo na poczatku mamy war pocz
    
    velocity = velocity_results[-1] + acceleration_results[-1] * t_step #new velocity
    theta = position_results[-1] + velocity * t_step #new position
    acc = acceleration(theta) #with new position (theta)
    
    velocity_results.append(velocity)
    position_results.append(theta)
    acceleration_results.append(acc)

T = list(map(lambda x: x * t_step, range(N+1))) #wiec tutaj o 1 wiecej


plt.plot(T, velocity_results)
plt.xlabel("T [s]")
plt.ylabel("V [rad]")
plt.title("$\\theta$ (s)")
plt.show()


plt.plot(T, position_results)
plt.xlabel("T [s]")
plt.ylabel("$\\theta$ [rad]")
plt.title("$\\theta$ (s)")
plt.grid()
plt.show()
```
