```python
import numpy as np
import matplotlib.pyplot as plt
 
N = 300 #liczba osob na wyspie
C = 15 #początkowa liczba kokosów przypadająca na mieszkańca
T = 300000 #liczba gier

# tworzę tablicę dla warunków początkowych --> 1 mieszkaniec - 15 kokosóW
x = np.ones(N) * C 

for t in range(T):
    # generuję losowo wybraną parę dwóch liczb (osób które grają w pnk)
    i, j = np.random.choice(N,2)
    
    # jeden warunek wystarczy 
    #               - wyłącznie przypadek dla którego się odejmuje kokos
    if x[j] > 0:    
        x[i] += 1 
        x[j] -= 1

    
#y = 1/(1+C) * np.exp(-x/C+0.5)

```

```python
_ = plt.hist(x, bins = 'auto', density =  True)

liczba_koko_na_mieszkanca = x.tolist()

y = []
xmin, xmax = min(x), max(x)
x_fit = np.linspace(xmin,xmax, num=201 )
y_fit = 1/(1+C) * np.exp(-x_fit/(C + 0.5))

_ = plt.plot(x_fit, y_fit, 'r')
#plt.show()
#plt.savefig('kokosy.png')
```
